# What is Machine Learning

&nbsp;


> Machine learning can be thought of as a way of **Pattern Recognition**

&nbsp;


We are given **DATA** as input. This **DATA** represents some **Pattern** *but the pattern is __unknown__*.

> The goal of machine learning then is to  
    >> * Detect *Patterns* and
    >> * Predict *Future data* using these patterns


***
___


__For example__ : if we are given a **DATASET** of **input,output** pairs

       D = {(input1,output1),......,(inputN,outputN)} 
   
       D = {(x1,y1),(x2,y2),(x3,y3),(x4,y4)}
     
&nbsp;
where x is the input 

      x = {x1,x2,x3,x4} = {1,2,3,4}
      
&nbsp;
and the y is the corresponding output

      y = {y1,y2,y3,y4} = {2,4,6,8}
 
 &nbsp;
 &nbsp;
 
> In simple terms, we try to find the **Pattern** by which the *input and output* are **related**.

     >> Which in this case is __y is twice the value of x__. 
     >>> We can say the relationship is **linear**.

&nbsp;
&nbsp;
>In mathematical terms, we try to find how y and x are related, 

     >> We try to find the function of x which determines y

         >>> ** y = `ƒ`(x) = 2 * x **
    

___

***
___


In Machine Learning, this underlying function `ƒ` is **unknown**.

>Goal of the Machine Learning Model is to find this __unknown function `ƒ`__.


___

***
___


## Classification or  regression

&nbsp;

As we have noted above, the Goal of learning is to
  >> Detect **Patterns**  +  Predict **Future data** using these patterns
 
We can describe these, using the below 2 terms respectively
>> **Function Approximation** + **Generalisation**


&nbsp;

Assuming ** y = `ƒ`(x) ** for some  _unknown function_ ** `ƒ` ** ..... we can say that the Goal of learning is

> To **Estimate a function** $\hat{f}$ , which represents the underlying data. 
>> In other words, _Estimate function_  ** $\hat{f}$ ** _such that_  .....  $\hat{f}$ `≈` ** `ƒ` **
     

> To **Make predictions** using this **Estimated Function** $\hat{f}$  
>> In other words, _Predict the value of y_ : ** $\hat{y}$ ** using the **Estimated Function $\hat{f}$ ** 
>>> $\hat{y}$  = $\hat{f}$(x)  ...... _where_ x is future(unseen) data
>>>> **Generalisation is prediction on unseen data**
