## <span style="color:blue">What is machine learning</span>



Machine learning can be thought of as a way of <span style="color:blue">**Pattern Recognition**</span>

<span style="color:blue">The Data represents some Pattern which is unknown.</span>

The basic idea is to use the data to 
    1. Detect patterns and
    2. Predict future data using these patterns


***
___


For example : if we are given a data set of **input,output** pairs

       D = {(input1,output1),......,(inputN,outputN)} 
   
       D = {(x1,y1),(x2,y2),(x3,y3),(x4,y4)}
     
where x is the input 

      x = {x1,x2,x3,x4} = {1,2,3,4}
      
and the y is the corresponding output

      y = {y1,y2,y3,y4} = {2,4,6,8}
 
>In simple terms, we try to find the **Pattern** by which the *input and output* are **related**.

     Which in this case is y is twice the value of x, and the relationship is linear.

>In mathematical terms, we try to find how y and x are related, 

     i.e the function of x which determines y

         y = f(x) = 2 * x
    

In Machine Learning, this underlying function f is <span style="color:red">**unknown**</span>.

>Goal of the Machine Learning Model is to find this unknown function f.


___

***
___


### <span style="color:blue">Classification or  regression</span>

Goal of learning is
>**Function approximation** + **generalisation**

Assume y = f(x) for some unknown function f

Goal of learning is

1. <span style="color:indigo">**Estimate function**</span> f^ ...... for a given labelled training set

2. Make predictions y^  = f^(x) ..... <span style="color:indigo">**Generalisation is prediction on unseen data**</span>


***
___


## <span style="color:blue">Machine learning : A probabilistic approach</span>



>Probability can be applied to any problem involving **uncertainty**


### <span style="color:green">Types of learning</span>


#### <span style="color:green">1. Supervised learning</span>


Maps input x to output y given a <span style="color:indigo">**labelled data set**</span> of input output pairs


      D = { (xi,yi) } where i = 1....n


 xi is a d dimensional vector of numbers that is features.
 
 yi is a response variable which can be
   1. Categorical yi belongs to {1....c} ......  This is <span style="color:indigo">**Classification**</span>
   2. Real valued yi belongs to R ....... That is <span style="color:indigo">**Regression**</span>
   

___

***
___


#### <span style="color:green">Using Probabilistic predictions</span>

P( y | x, D} ..... For a given input vector x and training set D

>Pick the best guess :  y^ = f^(x) = argmax{ P(y == c | x, D) }

<span style="color:blue">**Note**</span> if p(y^ | x, D) is far from 1.0 the answer should be don't know instead of giving incorrect guess.
