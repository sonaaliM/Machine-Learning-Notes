# Machine learning : A probabilistic approach



>Probability can be applied to any problem involving **uncertainty**



## The BIN Model

>Let us consider a BIN with RED and GREEN marbles

>>Suppose we pick a Sample of N marbles from this bin independently.

&nbsp;

>Let us define the below quantities


>>  **$\mu$**  $\rightarrow$ **Probability to pick a RED marble from the BIN**
&nbsp;

>>  **$\nu$** $\rightarrow$ **Fraction of RED marbles in the SAMPLE**


&nbsp;

>**NOTE**
>> We can think of the **SAMPLE** as the **DATASET** D 
>>> **$\nu$** $\rightarrow$ relates to the SAMPLE or DATASET

>> We can think of the **BIN** as the **Outside World**
>>> **$\mu$** $\rightarrow$ relates to the **Outside World**

&nbsp;

The question that arises is
> #### *Can we say anything about **$\mu$ (outside data)** after observing **$\nu$ (the dataset)** ?* 


>> ANSWER is **NO** ..... 
>>>*It is possible for the sample to be all GREEN but the BIN is mostly RED*

&nbsp;

The question then arises is
>> *Why do we trust polling which is based on similar logic?*
&nbsp;

>>> #### * Bcoz the BAD CASE is **POSSIBLE** but not **PROBABLE** *


&nbsp;

This we where we can make use of **Probability**

## Probability to the Rescue

>We consider here the **Hoeffding's Inequality** , which states that

>> #### *Most of the time **$\nu$** cannot be too far from **$\mu$** *


>>> ## **$\mathbb P$** $\big[$    |  $\nu$  -  $\mu$  |   $\gt$    **$\epsilon$**  $\big]$     $\leq$     **$\delta$**

>>>>  ###  **$\delta$**  = 2 $e^{-2 \epsilon^2 N}$
